```python
import pandas as pd
import json

# Ruta al archivo local
ruta_archivo = '../data/TelecomX_Data.json'
```

```python
# Abrimos el archivo y cargamos los datos como diccionario
with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
    datos = json.load(archivo)

# Convertimos a DataFrame
df = pd.DataFrame(datos)

# Mostramos las primeras filas
df.head()
```

¬°Perfecto! Ya cargaste correctamente el JSON y convertiste el contenido en un DataFrame.

Como vemos, las columnas customer, phone, internet y account contienen diccionarios anidados con m√°s informaci√≥n adentro. El siguiente paso es expandir esas columnas para poder trabajar con todos los datos como columnas planas.

```python
# Expandimos las columnas anidadas y las unimos al DataFrame principal
df_customer = df['customer'].apply(pd.Series)
df_phone = df['phone'].apply(pd.Series)
df_internet = df['internet'].apply(pd.Series)
df_account = df['account'].apply(pd.Series)

# Concatenamos todo en un nuevo DataFrame
df_expandido = pd.concat([df.drop(['customer', 'phone', 'internet', 'account'], axis=1),
                          df_customer, df_phone, df_internet, df_account], axis=1)

# Mostramos las primeras filas
df_expandido.head()
```

¬°Genial! Ya expandiste correctamente las columnas anidadas üéâ
Ahora ten√©s un DataFrame plano con todas las variables importantes como columnas, lo que facilita el an√°lisis.

‚úÖ Pr√≥ximo paso: Expandir la columna Charges
Esa columna todav√≠a contiene un diccionario con dos claves: "Monthly" y "Total". Vamos a separarla tambi√©n en dos columnas: MonthlyCharges y TotalCharges.

```python
# Expandimos la columna Charges
df_charges = df_expandido['Charges'].apply(pd.Series)

# Renombramos las columnas para mayor claridad
df_charges.columns = ['MonthlyCharges', 'TotalCharges']

# Unimos todo en un √∫nico DataFrame final
df_final = pd.concat([df_expandido.drop('Charges', axis=1), df_charges], axis=1)

# Mostramos las primeras filas del DataFrame final
df_final.head()
```

¬°Perfecto, Pablo! üéØ
Ya ten√©s tu DataFrame completamente expandido, limpio y listo para ser explorado.
Ahora vamos con el siguiente paso:

‚úÖ Paso 3: Exploraci√≥n b√°sica del dataset
Ejecut√° estos comandos para inspeccionar la estructura y tipo de datos:

```python
# Informaci√≥n general del DataFrame
df_final.info()

# Estad√≠sticas descriptivas
df_final.describe(include='all')
```

¬°Muy bien, Pablo! Ya detectamos informaci√≥n clave para el paso de detecci√≥n y manejo de inconsistencias. Ac√° te resumo lo m√°s relevante que encontramos:

‚úÖ Diagn√≥stico de inconsistencias
üìå 1. TotalCharges es object (texto)
Aunque deber√≠a ser num√©rico (float), aparece como tipo object. Esto suele pasar cuando hay espacios en blanco ('') o caracteres no num√©ricos. Lo solucionamos convirti√©ndolo a float.

üìå 2. Columnas categ√≥ricas: valores repetidos tipo "No internet service" o "No phone service"
Estas categor√≠as pueden tratarse como "No" para facilitar el an√°lisis, o mantenerse si se quiere conservar la distinci√≥n.

üîß Paso siguiente: limpiar TotalCharges y convertir a num√©rico
Ejecut√° este bloque para detectar y corregir los problemas:

```python
# Reemplazamos espacios vac√≠os por NaN
df_final['TotalCharges'] = df_final['TotalCharges'].replace(' ', pd.NA)

# Convertimos a num√©rico (float)
df_final['TotalCharges'] = pd.to_numeric(df_final['TotalCharges'])

# Verificamos si quedan valores nulos
print(df_final['TotalCharges'].isna().sum())

# (Opcional) Mostramos las filas con valores nulos si hay
df_final[df_final['TotalCharges'].isna()]
```

Excelente trabajo, Pablo. Ya descubrimos algo muy importante:

‚úÖ Resultado del an√°lisis:
üìå Hay 11 filas con TotalCharges nulo (NaN)
Esto sucede porque esas filas tienen tenure = 0, es decir, clientes nuevos que todav√≠a no han generado cargos totales.

üîß ¬øQu√© hacer con estos valores nulos?
üëâ Opci√≥n recomendada: eliminar esas 11 filas
Dado que representan menos del 0.2% del dataset (11 de 7267), lo m√°s pr√°ctico para tu an√°lisis es eliminarlas, ya que:

No tienen datos √∫tiles para calcular ‚ÄúTotalCharges‚Äù.

Su permanencia puede generar errores o sesgos al modelar.

‚úÖ Ejecut√° esto para eliminarlas:

```python
# Eliminamos filas con TotalCharges nulo
df_final = df_final.dropna(subset=['TotalCharges'])

# Confirmamos que ya no hay nulos
print(df_final['TotalCharges'].isna().sum())
```

¬°Perfecto, Pablo! üéØ
Ahora s√≠: no quedan valores nulos en TotalCharges, y el dataset est√° m√°s limpio.



‚úÖ Siguiente paso: normalizaci√≥n de variables categ√≥ricas
Algunas columnas tienen valores como "No internet service" o "No phone service", lo cual duplica significados (ej.: "No" vs. "No internet service").
Simplificarlos mejora el an√°lisis y la visualizaci√≥n.

```python
# Lista de columnas con valores como "No internet service"
columnas_a_normalizar = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Reemplazamos "No internet service" y "No phone service" por "No"
df_final[columnas_a_normalizar] = df_final[columnas_a_normalizar].replace('No internet service', 'No')
df_final['MultipleLines'] = df_final['MultipleLines'].replace('No phone service', 'No')
```

¬°Excelente, sigamos! üöÄ
