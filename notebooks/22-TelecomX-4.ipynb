```python
import pandas as pd
import json

# Ruta al archivo local
ruta_archivo = '../data/TelecomX_Data.json'
```

```python
# Abrimos el archivo y cargamos los datos como diccionario
with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
    datos = json.load(archivo)

# Convertimos a DataFrame
df = pd.DataFrame(datos)

# Mostramos las primeras filas
df.head()
```

¡Perfecto! Ya cargaste correctamente el JSON y convertiste el contenido en un DataFrame.

Como vemos, las columnas customer, phone, internet y account contienen diccionarios anidados con más información adentro. El siguiente paso es expandir esas columnas para poder trabajar con todos los datos como columnas planas.

```python
# Expandimos las columnas anidadas y las unimos al DataFrame principal
df_customer = df['customer'].apply(pd.Series)
df_phone = df['phone'].apply(pd.Series)
df_internet = df['internet'].apply(pd.Series)
df_account = df['account'].apply(pd.Series)

# Concatenamos todo en un nuevo DataFrame
df_expandido = pd.concat([df.drop(['customer', 'phone', 'internet', 'account'], axis=1),
                          df_customer, df_phone, df_internet, df_account], axis=1)

# Mostramos las primeras filas
df_expandido.head()
```

¡Genial! Ya expandiste correctamente las columnas anidadas 🎉
Ahora tenés un DataFrame plano con todas las variables importantes como columnas, lo que facilita el análisis.

✅ Próximo paso: Expandir la columna Charges
Esa columna todavía contiene un diccionario con dos claves: "Monthly" y "Total". Vamos a separarla también en dos columnas: MonthlyCharges y TotalCharges.

```python
# Expandimos la columna Charges
df_charges = df_expandido['Charges'].apply(pd.Series)

# Renombramos las columnas para mayor claridad
df_charges.columns = ['MonthlyCharges', 'TotalCharges']

# Unimos todo en un único DataFrame final
df_final = pd.concat([df_expandido.drop('Charges', axis=1), df_charges], axis=1)

# Mostramos las primeras filas del DataFrame final
df_final.head()
```

¡Perfecto, Pablo! 🎯
Ya tenés tu DataFrame completamente expandido, limpio y listo para ser explorado.
Ahora vamos con el siguiente paso:

✅ Paso 3: Exploración básica del dataset
Ejecutá estos comandos para inspeccionar la estructura y tipo de datos:

```python
# Información general del DataFrame
df_final.info()

# Estadísticas descriptivas
df_final.describe(include='all')
```

¡Muy bien, Pablo! Ya detectamos información clave para el paso de detección y manejo de inconsistencias. Acá te resumo lo más relevante que encontramos:

✅ Diagnóstico de inconsistencias
📌 1. TotalCharges es object (texto)
Aunque debería ser numérico (float), aparece como tipo object. Esto suele pasar cuando hay espacios en blanco ('') o caracteres no numéricos. Lo solucionamos convirtiéndolo a float.

📌 2. Columnas categóricas: valores repetidos tipo "No internet service" o "No phone service"
Estas categorías pueden tratarse como "No" para facilitar el análisis, o mantenerse si se quiere conservar la distinción.

🔧 Paso siguiente: limpiar TotalCharges y convertir a numérico
Ejecutá este bloque para detectar y corregir los problemas:

```python
# Reemplazamos espacios vacíos por NaN
df_final['TotalCharges'] = df_final['TotalCharges'].replace(' ', pd.NA)

# Convertimos a numérico (float)
df_final['TotalCharges'] = pd.to_numeric(df_final['TotalCharges'])

# Verificamos si quedan valores nulos
print(df_final['TotalCharges'].isna().sum())

# (Opcional) Mostramos las filas con valores nulos si hay
df_final[df_final['TotalCharges'].isna()]
```

Excelente trabajo, Pablo. Ya descubrimos algo muy importante:

✅ Resultado del análisis:
📌 Hay 11 filas con TotalCharges nulo (NaN)
Esto sucede porque esas filas tienen tenure = 0, es decir, clientes nuevos que todavía no han generado cargos totales.

🔧 ¿Qué hacer con estos valores nulos?
👉 Opción recomendada: eliminar esas 11 filas
Dado que representan menos del 0.2% del dataset (11 de 7267), lo más práctico para tu análisis es eliminarlas, ya que:

No tienen datos útiles para calcular “TotalCharges”.

Su permanencia puede generar errores o sesgos al modelar.

✅ Ejecutá esto para eliminarlas:

```python
# Eliminamos filas con TotalCharges nulo
df_final = df_final.dropna(subset=['TotalCharges'])

# Confirmamos que ya no hay nulos
print(df_final['TotalCharges'].isna().sum())
```

¡Perfecto, Pablo! 🎯
Ahora sí: no quedan valores nulos en TotalCharges, y el dataset está más limpio.



✅ Siguiente paso: normalización de variables categóricas
Algunas columnas tienen valores como "No internet service" o "No phone service", lo cual duplica significados (ej.: "No" vs. "No internet service").
Simplificarlos mejora el análisis y la visualización.

## 10 - Creación de Modelos
Descripción
Crea al menos dos modelos diferentes para predecir la cancelación de clientes.

Un modelo puede requerir normalización, como Regresión Logística o KNN.

El otro modelo puede no requerir normalización, como Árbol de Decisión o Random Forest.

💡 La decisión de aplicar o no la normalización depende de los modelos seleccionados. Ambos modelos pueden ser creados sin normalización, pero también es una opción combinar modelos con y sin normalización.

#### Justificación:

* Regresión Logística / KNN: Estos modelos son sensibles a la escala de los datos, por lo que la normalización es importante para que los coeficientes o las distancias se calculen correctamente.

* Árbol de Decisión / Random Forest: Estos modelos no dependen de la escala de los datos, por lo que no es necesario aplicar normalización.

Si decides normalizar los datos, deberías explicar cómo esta etapa asegura que los modelos basados en distancia o en optimización de parámetros no se vean sesgados por la magnitud de las variables.


## 11 - Evaluación de los Modelos
Descripción
Evalúa cada modelo utilizando las siguientes métricas:

* Exactitud (Acurácia)

* Precisión

* Recall

* F1-score

* Matriz de confusión

Después, realiza un análisis crítico y compara los modelos:

* ¿Cuál modelo tuvo el mejor desempeño?

* ¿Algún modelo presentó overfitting o underfitting? Si es así, considera las posibles causas y ajustes:

**Overfitting**: Cuando el modelo aprende demasiado sobre los datos de entrenamiento, perdiendo la capacidad de generalizar a nuevos datos. Considera reducir la complejidad del modelo o aumentar los datos de entrenamiento.

**Underfitting**: Cuando el modelo no captura bien las tendencias de los datos, lo que indica que es demasiado simple. Intenta aumentar la complejidad del modelo o ajustar sus parámetros.

## 📋 12 - Interpretación y Conclusiones
### Análisis de la Importancia de las Variables
Descripción
Después de elegir los modelos, realiza el análisis de las variables más relevantes para la predicción de la cancelación:

* Regresión Logística: Investiga los coeficientes de las variables, que muestran su contribución a la predicción de cancelación.

* KNN (K-Nearest Neighbors): Observa cómo los vecinos más cercanos influyen en la decisión de clasificación. Las variables más impactantes pueden ser aquellas que más contribuyen a la proximidad entre los puntos de datos.

* Random Forest: Utiliza la importancia de las variables proporcionada por el modelo. Random Forest calcula la importancia basándose en cómo cada variable contribuye a la reducción de la impureza durante las divisiones de los árboles.

* SVM (Support Vector Machine): En el SVM, las variables más relevantes son aquellas que influyen en la frontera de decisión entre las clases. Puedes analizar los coeficientes de los vectores de soporte para entender qué variables tienen mayor impacto.

* Otros Modelos: Dependiendo del modelo elegido, considera el análisis de métricas específicas para comprender la relevancia de las variables. Por ejemplo, coeficientes en modelos lineales, pesos en redes neuronales, o la importancia relativa en técnicas de boosting (como XGBoost).
