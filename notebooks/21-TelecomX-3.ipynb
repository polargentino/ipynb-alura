## üîß Paso 5: Desanidar columnas y corregir inconsistencias
### ‚úÖ Prompt 5

"""
Desanid√° las columnas 'customer', 'phone', 'internet' y 'account', que contienen diccionarios.
Convert√≠ cada clave de esos diccionarios en nuevas columnas.
Luego, correg√≠ la columna 'Churn' reemplazando los valores vac√≠os ('') por NaN.
"""


```python
# Importar numpy para usar NaN
import numpy as np

# 1. Expandir columnas anidadas
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

# 2. Unir todo en un solo DataFrame plano
df_final = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)

# 3. Reemplazar valores vac√≠os en Churn por NaN
df_final['Churn'] = df_final['Churn'].replace('', np.nan)

# Verificar el resultado
df_final.head()
```

```python
# Paso 0: Importar librer√≠as
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 1: Cargar JSON desde archivo
with open('../data/TelecomX_Data.json', 'r') as f:
    datos_json = json.load(f)

df = pd.DataFrame(datos_json)

# Paso 2: Desanidar columnas
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

df_final = pd.concat([df[['customerID', 'Churn']], df_customer, df_phone, df_internet, df_account], axis=1)

# Paso 3: Limpieza b√°sica
df_final['Churn'] = df_final['Churn'].replace('', np.nan)

# Paso 4: Renombrar columnas con puntos o espacios
df_final.columns = df_final.columns.str.strip().str.replace(' ', '_').str.replace('.', '_')

# Paso 5: Conversi√≥n a num√©rico
df_final['Charges_Monthly'] = pd.to_numeric(df_final['Charges_Monthly'], errors='coerce')
df_final['Charges_Total'] = pd.to_numeric(df_final['Charges_Total'], errors='coerce')
df_final['tenure'] = pd.to_numeric(df_final['tenure'], errors='coerce')

# Paso 6: Crear columna diaria
df_final['Cuentas_Diarias'] = df_final['Charges_Monthly'] / 30

# Paso 7: Reemplazar Yes/No
df_final.replace({'Yes': 1, 'No': 0}, inplace=True)
```
