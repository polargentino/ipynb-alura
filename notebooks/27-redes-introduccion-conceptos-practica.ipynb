# 10 Traceroute
¡Hola! En esta clase, aprendimos sobre el comando traceroute (o tracert en Windows) que nos permite rastrear la ruta que siguen los paquetes de datos desde nuestro dispositivo hasta un servidor de destino en Internet.Aquí hay algunos puntos clave:
Función del comando: traceroute nos ayuda a identificar los nodos (enrutadores) por los que pasan nuestros datos y a detectar posibles problemas de conexión en el camino.
Funcionamiento: El comando envía paquetes ICMP a cada nodo y registra el tiempo de respuesta. Esto nos permite ver la ruta que siguen los paquetes y la latencia en cada salto.
Saltos: Cada nodo por el que pasa el paquete se considera un "salto". El traceroute muestra una lista de estos saltos, junto con la dirección IP y el tiempo de respuesta de cada uno.
Tiempo de espera agotado: En algunos casos, podemos ver un mensaje de "tiempo de espera agotado" (tres asteriscos) en lugar de la información de un nodo. Esto puede ocurrir porque el administrador del dispositivo ha deshabilitado las respuestas ICMP por motivos de seguridad o para evitar la congestión del tráfico.
Número máximo de saltos: El traceroute tiene un límite máximo de saltos (30 por defecto) para evitar bucles infinitos en la red.
Congestión: La congestión en la red puede afectar el tiempo de respuesta de los nodos.

En resumen, traceroute es una herramienta útil para diagnosticar problemas de red y comprender cómo se enrutan los datos en Internet. ¿Te gustaría saber algo más sobre algún punto en específico?

Ello porque, en redes de computadores tenemos lo que se conoce como red no determinística, es decir, un paquete de información no será necesariamente transferido por la misma ruta del anterior con el mismo intervalo de tiempo.
