# 10 Traceroute
¡Hola! En esta clase, aprendimos sobre el comando traceroute (o tracert en Windows) que nos permite rastrear la ruta que siguen los paquetes de datos desde nuestro dispositivo hasta un servidor de destino en Internet.Aquí hay algunos puntos clave:
Función del comando: traceroute nos ayuda a identificar los nodos (enrutadores) por los que pasan nuestros datos y a detectar posibles problemas de conexión en el camino.
Funcionamiento: El comando envía paquetes ICMP a cada nodo y registra el tiempo de respuesta. Esto nos permite ver la ruta que siguen los paquetes y la latencia en cada salto.
Saltos: Cada nodo por el que pasa el paquete se considera un "salto". El traceroute muestra una lista de estos saltos, junto con la dirección IP y el tiempo de respuesta de cada uno.
Tiempo de espera agotado: En algunos casos, podemos ver un mensaje de "tiempo de espera agotado" (tres asteriscos) en lugar de la información de un nodo. Esto puede ocurrir porque el administrador del dispositivo ha deshabilitado las respuestas ICMP por motivos de seguridad o para evitar la congestión del tráfico.
Número máximo de saltos: El traceroute tiene un límite máximo de saltos (30 por defecto) para evitar bucles infinitos en la red.
Congestión: La congestión en la red puede afectar el tiempo de respuesta de los nodos.

En resumen, traceroute es una herramienta útil para diagnosticar problemas de red y comprender cómo se enrutan los datos en Internet. ¿Te gustaría saber algo más sobre algún punto en específico?

# 07 Cables de conexión
¡Hola! En esta clase, Álvaro nos explica sobre los cables de conexión, específicamente el cable UTP (par trenzado sin blindaje) y su conector RJ45, muy común para conectar dispositivos a Internet en casa.Aquí están los puntos clave:

Estándar 568: Define cómo se cablean los terminales del conector RJ45. El estándar 568 consiste en que los terminales del conector RJ45 van cableados de la siguiente forma: el terminal 1 va a tener el blanco con verde, el terminal 2, el verde, el terminal 3, blanco con naranja, el cuatro, el azul, el 5, el blanco con azul, el 6, el naranja, el 7, el blanco con marrón y el 8 el marrón.

Cable Directo (568A): La configuración de colores es la misma en ambos extremos del cable. Se usa para conectar dispositivos diferentes, como un computador a un hub.

Cable Cruzado (568A y 568B): Un extremo del cable tiene la configuración 568A y el otro la 568B. Se usa para conectar dispositivos iguales, como dos computadores. Esto evita colisiones, ya que los terminales de transmisión y recepción se invierten.

Cuándo usar cada uno:
Directo: Switch a router, switch a computador, switch a servidor, hub a computador, hub a servidor.
Cruzado: Switch a switch, switch a hub, hub a hub, router a router, computador a router, computador a computador.

# 3 - Wireshark, Switches y avances
## 01 - Limitación del Hub
¡Hola! En esta clase, se explica cómo funciona la comunicación entre dispositivos en una red utilizando un hub, centrándose en los protocolos ARP e ICMP.Aquí tienes un resumen de los puntos clave:
Simulación en Packet Tracer: Se utiliza Packet Tracer para simular una red con tres computadoras conectadas a través de un hub.

Protocolo ARP (Address Resolution Protocol):
Antes de enviar un mensaje ICMP (ping), se utiliza ARP para descubrir la dirección MAC correspondiente a una dirección IP.
El PC2 envía una solicitud ARP al hub para encontrar la dirección MAC del PC0 (192.168.1.2).
El hub difunde esta solicitud a todos los dispositivos conectados.
El PC0 responde con su dirección MAC, que es enviada de vuelta al PC2 a través del hub.


Limitaciones del Hub:
El hub no aprende ni mantiene una tabla de direcciones MAC. Simplemente reenvía todo el tráfico que recibe a todos los puertos.

Esto crea una vulnerabilidad de seguridad, ya que todos los dispositivos en la red pueden ver la información que se transmite.
También puede causar congestión en la red, especialmente con grandes cantidades de datos.


Protocolo ICMP (Internet Control Message Protocol):
Después de resolver la dirección MAC con ARP, se envía un paquete ICMP (ping) al PC0.
El hub difunde este paquete a todos los dispositivos, y el PC0 responde.
El PC2 recibe la respuesta y muestra los resultados del ping en el command prompt.
