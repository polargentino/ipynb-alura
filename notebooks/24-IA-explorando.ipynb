# Conociendo Google Gemini 
¡Hola! En esta clase, exploramos Google Gemini, un modelo de lenguaje (LLM) desarrollado por Google, como una alternativa a ChatGPT.Aquí tienes un resumen de los puntos clave que cubrimos:
Acceso y Similitudes: Gemini se puede acceder a través de gemini.google.com con una cuenta de Gmail, y su interfaz es similar a la de ChatGPT.
Conexión con Google: Una ventaja de Gemini es su conexión directa con el buscador de Google, lo que le permite acceder a información actualizada en tiempo real.
Ejemplos de uso: Probamos Gemini con los mismos prompts que usamos en ChatGPT, como buscar hoteles en Roma que sean amigables para niños y crear una publicación de blog.
Comparación: Observamos que Gemini puede tener ventajas en la búsqueda de información actualizada, pero ChatGPT puede ser más preciso en la interpretación de los prompts.
Limitaciones: Gemini tuvo problemas para encontrar hoteles con sauna y tendía a usar listas incluso cuando se le pedía que no lo hiciera.
Tono informal: Intentamos reescribir una publicación de blog con un tono más informal, y Gemini utilizó jerga coloquial para lograrlo.

En resumen, Gemini es una herramienta poderosa con acceso a la información de Google, pero puede tener algunas limitaciones en comparación con ChatGPT en términos de precisión y cumplimiento de instrucciones específicas.¿Te gustaría que profundicemos en algún aspecto en particular o que hagamos algún ejercicio para practicar el uso de Gemini?

#  06 Mejorando el contexto
 Como vimos en clase, es importante que el comando enviado a la IA generativa tenga un contexto bien definido. Existen diferentes técnicas ya estudiadas para elaborar buenos prompts, y una de las más utilizadas es el uso de ejemplos.

¿Cuál es el beneficio de usar ejemplos específicos en un prompt al interactuar con ChatGPT o Gemini?

Alternativa incorreta
Aumenta la ambigüedad de la respuesta.


¡Muy al contrario! Al proporcionar ejemplos, aumentamos la posibilidad de obtener la respuesta deseada.

Alternativa incorreta
Reduce la complejidad del prompt.


El prompt se vuelve más complejo cuando añadimos ejemplos.

Alternativa incorreta
Garantiza que el modelo proporcione respuestas más precisas y relevantes.


Los ejemplos enriquecen el contexto definido en el prompt y el texto generado por la IA tendrá el tono de voz y formato basado en los ejemplos proporcionados.

Alternativa incorreta
Reduce la necesidad de proporcionar contexto adicional.


El contexto sigue siendo importante. Lo que los ejemplos hacen es enriquecer el contexto y mostrar el formato deseado de la respuesta.
