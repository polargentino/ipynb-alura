````python
# Aplanar la columna 'customer'
customer_df = pd.json_normalize(df['customer'])
df = pd.concat([df.drop('customer', axis=1), customer_df], axis=1)

# Aplanar la columna 'phone'
phone_df = pd.json_normalize(df['phone'])
df = pd.concat([df.drop('phone', axis=1), phone_df], axis=1)

# Aplanar la columna 'internet'
internet_df = pd.json_normalize(df['internet'])
df = pd.concat([df.drop('internet', axis=1), internet_df], axis=1)

# Aplanar la columna 'account'
account_df = pd.json_normalize(df['account'])
df = pd.concat([df.drop('account', axis=1), account_df], axis=1)

print("DataFrame después de aplanar todas las columnas anidadas:")
print(df.head())
print(f"\nNúmero de columnas después de aplanar: {df.shape[1]}")
print("\nTipos de datos actuales de las columnas:")
print(df.dtypes)
````

## 🧹 Limpieza y Conversión de Charges.Total
Vamos a inspeccionar y corregir la columna Charges.Total. Es común que los valores vacíos o no numéricos en esta columna representen clientes nuevos que aún no han acumulado cargos totales. En estos casos, suelen ser cero o un valor pequeño.



````python
# Inspeccionar valores únicos en Charges.Total para entender el problema
print("Valores únicos en 'Charges.Total' antes de la conversión (primeros 20):")
print(df['Charges.Total'].unique()[:20])

# Reemplazar espacios vacíos o ' ' con NaN y luego convertir a numérico
# El argumento errors='coerce' convertirá cualquier valor no numérico en NaN
df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')

# Verificar si hay valores NaN después de la conversión
print(f"\nNúmero de valores NaN en 'Charges.Total' después de la conversión: {df['Charges.Total'].isnull().sum()}")

# Llenar los valores NaN (por ejemplo, con 0, la media, o la mediana).
# Para cargos totales, es razonable asumir 0 para nuevos clientes sin cargos registrados.
df['Charges.Total'] = df['Charges.Total'].fillna(0) # Opcional: df['Charges.Total'].mean() para la media

# Verificar el tipo de dato después de la limpieza
print(f"Tipo de dato de 'Charges.Total' después de la limpieza: {df['Charges.Total'].dtype}")

print("\nPrimeras filas del DataFrame con 'Charges.Total' limpio:")
print(df.head())
````

````python
# Identificar columnas categóricas (tipo 'object')
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Excluir la columna objetivo 'Churn' si está en las categóricas para no encodificarla aún
if 'Churn' in categorical_cols:
    categorical_cols.remove('Churn')

print(f"\nColumnas categóricas a codificar: {categorical_cols}")

# Aplicar One-Hot Encoding
# drop_first=True evita la trampa de las variables dummy (multicolinealidad)
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("\nDataFrame después de One-Hot Encoding:")
print(df.head())
print(f"\nNúmero de columnas después de encoding: {df.shape[1]}")
print("\nTipos de datos de las columnas después de encoding:")
# Esto nos dará un resumen de cuántas columnas hay por cada tipo de dato
print(df.dtypes.value_counts())
````

### 📊 Análisis de Desbalanceo de Clases y Conversión de 'Churn'
Evaluar la proporción de clientes que cancelaron versus los que no es fundamental, ya que un desbalanceo significativo puede sesgar tus modelos. Además, transformaremos "Yes"/"No" en 1/0.

````python
# Calcular la proporción de la clase 'Churn'
churn_counts = df['Churn'].value_counts()
total_customers = df.shape[0]

print("Conteo de clientes por estado de Churn:")
print(churn_counts)

print("\nProporción de Churn:")
print(f"No Churn: {churn_counts.get('No', 0) / total_customers:.2%}")
print(f"Churn: {churn_counts.get('Yes', 0) / total_customers:.2%}")

# Convertir la columna 'Churn' a formato numérico (0 y 1)
# Asumiendo 'Yes' para Churn (1) y 'No' para No Churn (0)
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

print("\nProporción numérica de Churn (0=No, 1=Yes):")
print(df['Churn'].value_counts(normalize=True))

if churn_counts.get('Yes', 0) / total_customers < 0.3 or churn_counts.get('Yes', 0) / total_customers > 0.7:
    print("\n¡Advertencia! Existe un **desbalanceo significativo** en las clases de Churn.")
    print("Considera aplicar técnicas de balanceo como **SMOTE** si los modelos iniciales no rinden bien, especialmente en métricas como Recall para la clase 'Churn'.")
else:
    print("\nLas clases de Churn parecen estar razonablemente balanceadas.")

print("\nPrimeras filas del DataFrame con 'Churn' convertido a numérico:")
print(df.head())
print(f"\nTipo de dato de 'Churn' después de la conversión: {df['Churn'].dtype}")
````

¡Excelente! La columna Churn ha sido correctamente convertida a un tipo numérico (float64), con 1 para "Yes" (canceló) y 0 para "No" (no canceló). También has confirmado el desbalanceo significativo en las clases, lo cual es una observación crucial para la fase de modelado.

````python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# --- Modelo 1: Regresión Logística ---
# Usamos el solver 'liblinear' que es bueno para datasets pequeños y medianos,
# y maneja bien la regularización (aunque aquí no la ajustamos explícitamente, es un buen default).
# random_state asegura la reproducibilidad.
log_reg_model = LogisticRegression(random_state=42, solver='liblinear')
log_reg_model.fit(X_train, y_train)

print("\n--- Modelo de Regresión Logística entrenado ---")

# --- Modelo 2: Random Forest ---
# n_estimators: número de árboles en el bosque (un valor común es 100).
# random_state asegura la reproducibilidad.
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(X_train, y_train)

print("--- Modelo Random Forest entrenado ---")
````

### 📊 Evaluación y Comparación de Modelos
Vamos a utilizar las métricas estándar de clasificación para evaluar el desempeño de cada modelo en el conjunto de prueba (X_test, y_test).

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # Importamos numpy para el slicing de probabilidades si es necesario

def evaluate_model(model, X_test, y_test, model_name):
    """
    Evalúa un modelo de clasificación y muestra sus métricas clave,
    incluyendo la matriz de confusión.
    """
    y_pred = model.predict(X_test)
    # Algunas modelos tienen predict_proba, otros no (ej. algunos con ensemble/stacking)
    # Se usa try-except para manejar esto si fuera necesario, pero para LR y RF siempre existe.
    # y_proba = model.predict_proba(X_test)[:, 1] # Probabilidades para la clase positiva

    print(f"\n--- Métricas para: {model_name} ---")
    print(f"Exactitud (Accuracy): {accuracy_score(y_test, y_pred):.4f}")
    # Precision, Recall y F1-score para la clase positiva (Churn=1)
    print(f"Precisión (Clase 1 - Churn): {precision_score(y_test, y_pred):.4f}")
    print(f"Recall (Clase 1 - Churn): {recall_score(y_test, y_pred):.4f}")
    print(f"F1-Score (Clase 1 - Churn): {f1_score(y_test, y_pred):.4f}")

    cm = confusion_matrix(y_test, y_pred)
    print("\nMatriz de Confusión:")
    print(cm)

    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['No Churn (0)', 'Churn (1)'], yticklabels=['No Churn (0)', 'Churn (1)'])
    plt.xlabel('Predicción')
    plt.ylabel('Real')
    plt.title(f'Matriz de Confusión: {model_name}')
    plt.show()

# Evaluar el modelo de Regresión Logística
evaluate_model(log_reg_model, X_test, y_test, "Regresión Logística")

# Evaluar el modelo Random Forest
evaluate_model(rf_model, X_test, y_test, "Random Forest")
````

# Resumen Ejecutivo: Análisis Predictivo de Cancelación (Churn) - Telecom X

**Objetivo:** Desarrollar y evaluar modelos de Machine Learning para prever la cancelación de clientes y proponer estrategias de retención proactivas.

**Metodología Clave:**
* **Preprocesamiento de Datos:** Limpieza, aplanamiento de estructuras anidadas, codificación de variables categóricas y estandarización de datos numéricos. Se gestionaron 224 valores `NaN` en la variable objetivo `Churn`.
* **Modelos:** Entrenamiento de **Regresión Logística** y **Random Forest**.
* **Evaluación:** Comparación basada en Exactitud, Precisión, Recall y F1-Score, priorizando el **Recall para la clase 'Churn'** (identificación de clientes en riesgo) debido al desbalanceo (26.5% de Churn).

**Resultados del Modelo:**
* Ambos modelos mostraron un rendimiento comparable con ~79% de exactitud.
* **La Regresión Logística fue marginalmente superior** en la detección de `Churn` (Recall: 0.5214 vs. 0.4920 de Random Forest; F1-Score: 0.5718 vs. 0.5601), identificando a más clientes que realmente cancelaron.

**Factores Clave de Cancelación (Drivers de Churn):**
El análisis de coeficientes de la Regresión Logística reveló los siguientes factores críticos:

1.  **`tenure` (Tiempo de Contrato):** **Principal factor de retención.** Mayor permanencia reduce drásticamente la probabilidad de Churn.
2.  **`InternetService_Fiber optic`:** **Alto riesgo de Churn.** Clientes con fibra óptica son significativamente más propensos a cancelar.
3.  **`Contract_Two year`:** **Fuerte factor de retención.** Contratos a largo plazo reducen el Churn.
4.  **`Charges.Total`:** Mayor gasto total acumulado **aumenta la probabilidad de Churn.**
5.  **`PaymentMethod_Electronic check`:** Este método de pago **aumenta la probabilidad de Churn.**
6.  **`Charges.Monthly`:** Mayor cargo mensual **reduce la probabilidad de Churn** (interesante, podría indicar satisfacción con planes de mayor valor).
7.  **Servicios de Valor Añadido (ej. Soporte Técnico, Seguridad Online):** La presencia de estos servicios **reduce el Churn.**

**Estrategias de Retención Propuestas:**

* **Foco en Clientes Nuevos:** Programas de "onboarding" intensivos y seguimiento proactivo en los primeros meses (`tenure` bajo).
* **Optimización de Fibra Óptica:** Investigar y resolver problemas de satisfacción en el servicio de fibra óptica (calidad, precio, soporte).
* **Fomento de Contratos a Largo Plazo:** Ofrecer incentivos atractivos para contratos de 1 y 2 años.
* **Incentivos para Métodos de Pago más "Estables":** Promover el cambio de cheque electrónico a otros métodos de pago automáticos.
* **Promoción de Servicios de Valor Añadido:** Destacar y fomentar la adopción de servicios como soporte técnico y seguridad online.

**Conclusión:**
Hemos establecido una base sólida para la predicción de Churn en Telecom X. La Regresión Logística, por su capacidad de identificar a más clientes en riesgo, es el modelo inicial recomendado.
